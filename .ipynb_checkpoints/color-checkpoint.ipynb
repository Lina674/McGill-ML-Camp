```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv("combined-data.csv")
ds
```

```python
ds["colour"] = ds["colour"].str.lower()
ds
```

```python
# print(ds.to_string()) #221-260
ds.loc[221:260, ['R', 'G', 'B']] = ds.loc[221:260, ['R', 'G', 'B']].apply(lambda x: x*255)
ds
```

```python
ds["colour"] = ds.loc[:, ["colour"]].apply(lambda name: name.str.split().str[-1])
ds
```

```python
v_c = ds["colour"].value_counts()
v_v = v_c[v_c > 7].index
df = ds[ds["colour"].isin(v_v)]
# print(df.to_string())
df
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

np.random.seed(6)

X, y = df[['R', 'G','B']], df['colour']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

knn = KNeighborsClassifier(1)
knn.fit(X_train, y_train)

y_predicted = knn.predict(X_test)
accuracy_score(y_test, y_predicted)
```

```python
def best_k():
    max, best_i = 0, 0
    for i in range(1, 50):
        knn = KNeighborsClassifier(i)
        knn.fit(X_train, y_train)

        y_predicted = knn.predict(X_test)
        grade = accuracy_score(y_test, y_predicted)
        if grade > max:
            max = grade
            best_i = i
    return best_i
```

```python
k = best_k()
print(k)

knn = KNeighborsClassifier(k)
knn.fit(X_train, y_train)

y_predicted = knn.predict(X_test)
accuracy_score(y_test, y_predicted)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_predicted = dt.predict(X_test)
print(accuracy_score(y_test, y_predicted))
```

```python
from sklearn.ensemble import RandomForestClassifier
np.random.seed(0)
rdt = RandomForestClassifier(n_estimators=59)
rdt.fit(X_train, y_train)
y_predicted = rdt.predict(X_test)
accuracy_score(y_test, y_predicted)
```

```python
def best_trees(rdt):
    max, best_n_tree = 0, 0
    for i in range(50,300):
        rdt = RandomForestClassifier(n_estimators=i)
        rdt.fit(X_train, y_train)
        y_predicted = rdt.predict(X_test)
        a_s = accuracy_score(y_test, y_predicted)
        if a_s > max:
            max = a_s
            best_n_tree = i
    return best_n_tree
```
